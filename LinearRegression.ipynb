import numpy as np
import pandas as pd

```python
import numpy as npimport matplotlib.pyplot as plt
```

```python
def grad(X, Y, theta, L, epochs):
    n = len(Y)
    history = []  
    for i in range(epochs):
        pred = X.dot(theta)
        errors = Y - pred
        gradients = (-2/n) * X.T.dot(errors)
        theta = theta - L * gradients
        cost = (1/n) * np.sum(errors**2)
        history.append(cost)
    
    return theta, history,cost
```

```python
def fit(X,Y,theta, lr,epochs):
    theta,history,loss = grad(X,Y,theta,lr,epochs)
    return theta,history,loss
    
```

```python
np.random.seed(42) 
X = 2*np.random.rand(100, 1)
Y = 2 + 7 * X + np.random.randn(100, 1)
```

```python
theta = np.zeros((2, 1))
lr=0.1
epochs = 1000
X_b = np.c_[np.ones((100, 1)), X] 
theta ,history,loss = fit(X_b,Y,theta,lr,epochs)
print(f"Theta: {theta}")
print("*"*20)
print(f"loss: {loss}")
```

```python
plt.scatter(X, Y)
plt.plot(X, X_b.dot(theta), color='red') 
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
